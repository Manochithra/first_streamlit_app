```python
# *args >> accept any number of positional arguments
def func(*args):
    print(args, type(args))
func()
func(1,2)
func(1,2,3,4,5,6,7)
```

```python
# *args >> accept any number of positional arguments
# **kwargs >> accept any number of keyword arguments
def func(*args, **kwargs):
    print(args, type(args))
    print(kwargs, type(kwargs))
func()
print("===========================")
func(a="a",b="b")
print("===========================")
func(1,2, a=100,z =200)
print("===========================")
func(1,2,3,4,5,6,7)
```

```python
# default arguments
def add(a,b=0):
    print(f">> add>> a={a} b={b} a+b = {a+b}")
    return a+b
# positional arguments
print(add(5,2))
print(add(2,5))
# keyword arguments
print(add(b=2,a=5))

print(add(5))
```

```python
#               lambda inp:out
add_lambda = lambda a,b=0: a+b
# positional arguments
print(add_lambda(5,2))
print(add_lambda(2,5))
# keyword arguments
print(add_lambda(b=2,a=5))

print(add_lambda(5))
```

```python
print(type(add))
print(type(add_lambda))
```

```python
random_func = lambda *args, **kwargs: {"args":args, "kwargs":kwargs}

print(random_func())
print("===========================")
print(random_func(a="a",b="b"))
print("===========================")
print(random_func(1,2, a=100,z =200))
print("===========================")
print(random_func(1,2,3,4,5,6,7))
```
